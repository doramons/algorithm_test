```python
def d(n): # n과 n의 각 자리수를 더하는 함수 d(n) 선언
    answer = 0
    for i in list(str(n)):
        answer += int(i)
    return answer +n
```

```python
d(1) # d(n)이 제대로 실행하는지 확인
```

```python
arr = set(range(1,10001)) # 10000이하의 숫자 모음
not_self = [] # 그 중 셀프가 아닌 숫자를 넣을 리스트
for i in range(1,10001): # 1부터 10000까지의 숫자를 d(n) 에 넣어 리턴되는 값은 셀프 넘버가 아니므로 not_self 리스트에 넣음
    not_self.append(d(i))
```

```python
not_self # not_self 
```

```python
arr
```

```python
self_number = list(set(arr) - set(not_self))
```

```python
for i in sorted(self_number):
    print(i)
```
